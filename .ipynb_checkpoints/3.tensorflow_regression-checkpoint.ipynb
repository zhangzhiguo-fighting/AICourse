```python
'''
This script works on TensorFlow 1.x : )
'''
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
%pylab inline
print(tf.__version__)
```

```python
#绘图
def plot_points(x, y, title_name):
    plt.title(title_name)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.scatter(x, y) #在图上显示x，y点
    plt.show()
```

```python
def plot_line(w, b, title_name):
    plt.title(title_name)
    plt.xlabel('x')
    plt.ylabel('y')
    x = np.linspace(0.0, 2.0, num=100)
    y = w * x + b
    plt.plot(x, y)
    plt.show()
```

```python
train_X = np.array([30.0, 40.0, 60.0, 80.0, 100.0, 120.0, 140.0])
train_Y = np.array([320.0, 360.0, 400.0, 450.0, 490.0, 550.0, 580.0])
#归一化
train_X /= 100.0
train_Y /= 100.0

plot_points(train_X, train_Y, title_name='Trainning Points')
```

```python
#计算图
X = tf.placeholder(tf.float32, None)
y = tf.placeholder(tf.float32, None)
#需要训练的变量（参数）
W = tf.Variable(np.random.randn(), name='weight') #加上name，方便自己更容易的找到
b = tf.Variable(np.random.randn(), name='bias')
n_samples = train_X.shape[0]
y_pred = tf.add(tf.multiply(W, X), b)
#print(w)
#print(b)
#print(y_pred)
```

```python
#对训练效果进行评测
cost = tf.reduce_sum(tf.pow((y_pred - y), 2)) / (2 * n_samples)
learning_rate = 0.01
optimizer = tf.train.GradientDescentOptimizer(learning_rate).minimize(cost) #告诉框架我们想用学习率去做什么
```

```python
training_epoch = 1000
display_step = 50
```

```python
with tf.Session() as sess:
    sess.run(tf.initialize_all_variables()) #对所有的参数进行一次初始化
    for epoch in range(training_epoch):
        for (x_train, y_train) in zip(train_X, train_Y): #打包数据
            _, loss, current_W, current_b = sess.run([optimizer, cost, W, b], feed_dict={X: x_train, y: y_train})
            if epoch % display_step == 0:
                print('Iteration: %04d | Loss: %.6f | W: %.6f | b: %.6f'
                       % (epoch + 1, loss, current_W, current_b))
                plot_line(current_W, current_b, 'Model Parameter')
    print('Training loss: %.6f | W: %.6f | b: %.6f' % (loss, sess.run(W), sess.run(b)))
```
